```python
import glob
import mysql.connector as connector
data_dir = 'stock_dfs'
config = {
   'user': 'jameschl',
   'password': '123456',
   'database': 'yahoo',
   'host': '127.0.0.1'
}
conn = connector.connect(**config)
c = conn.cursor()
def create_prod_table(ticker, c):
    table_name = ticker + 'Prod'
    c.execute('create table ' + table_name + 
              """(Date date, 
               Open float, 
               High float, 
               Low float, 
               Close float, 
               Adj_Close float, 
               Volume float )""", )
def alter_prod_table(ticker, c):
    prod_table_name = ticker + 'Prod'
    staging_table_name = ticker + 'Staging'
    c.execute('alter table ' + prod_table_name +
              """ add constraint symbol_day unique (Date)
              """, )
    
    c.execute('insert ignore into ' + prod_table_name +
              '(select * from ' + staging_table_name + ')'
              , )
    
def insert_staging_row(filename, ticker, c):
    print('insert row')
    delimiter = r','
    dateString = r'%d-%b-%Y'
    table_name = ticker + 'Staging'
    c.execute("Load data local infile %s into table " + table_name + 
              " fields terminated by %s ignore 1 lines" +
             "(Date,Open,High,Low,Close,Adj_Close,Volume)",
             (filename, delimiter))
    print('load file', filename, 'into', table_name)
```

```python
def insert_prod_db():
    for f in glob.glob(data_dir + '/*'):
        if (f.endswith('.csv')):
            ticker = f.split('/')[1].split('.')[0]
            print('create prod ticker', ticker)
            try:
                create_prod_table(ticker, c)
                alter_prod_table(ticker, c)
            except Exception as e:
                print(e)
            finally:
                conn.commit()
    c.close()
    conn.close()
```
